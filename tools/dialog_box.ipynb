```python
from PIL import Image, ImageDraw

# Функция для рисования диалогового окна
def draw_dialog_box(min_w, msg_frame_h, quest_frame_h):
    frame_w = 50  # Минимальная ширина окна
    frame_h = 50

    # Создаем новое изображение с альфа-каналом (прозрачным фоном)
    img = Image.new('RGBA', (frame_w, frame_h), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    # Первый прямоугольник (внешний)
    draw.rounded_rectangle(
        [(0, 0), (frame_w, frame_h)], 
        radius=10, 
        fill=(0, 0, 0)
    )
    
    # Второй прямоугольник (синий фон)
    draw.rounded_rectangle(
        [(1, 1), (frame_w - 2, frame_h - 2)], 
        radius=10, 
        fill=(209, 209, 209)
    )
    
    
    return img

# Параметры диалогового окна
min_w = 300
msg_frame_h = 50
quest_frame_h = 50

# Рисуем и сохраняем изображение
dialog_box_image = draw_dialog_box(min_w, msg_frame_h, quest_frame_h)
dialog_box_image.save('dialog_box.png')
```

```python
def drawRoundRect(draw, x, y, width, height, radius, fill):
    """
    Рисует закругленный прямоугольник на объекте draw.
    
    :param draw: Объект ImageDraw.Draw.
    :param x: Координата X верхнего левого угла прямоугольника.
    :param y: Координата Y верхнего левого угла прямоугольника.
    :param width: Ширина прямоугольника.
    :param height: Высота прямоугольника.
    :param radius: Радиус закругления углов.
    """
    # Используем встроенную функцию rounded_rectangle для рисования закругленного прямоугольника
    draw.rounded_rectangle(
        [x, y, x + width, y + height], 
        radius=radius, 
        outline=None, 
        fill=fill
    )
```
